# 04 - Generate journals JIF-Q1 file

```python
# Imports iniciais
import pandas as pd

# Pacote para ignorar mensagem de warning
import warnings
warnings.filterwarnings('ignore')
```

```python
df_jif = pd.read_excel('../../data/raw/JCR_JournalResults_09_2024.xlsx', sheet_name=1)
```

```python
df_jif.head()
```

```python
# Remover espaços em branco e converter para maiúsculas os dados da coluna 'JIF Quartile'
df_jif['JIF Quartile'] = df_jif['JIF Quartile'].str.strip().str.upper()

# Filtrar os períodicos quartil 1
df_jif_q1 = df_jif[df_jif['JIF Quartile'] == 'Q1']
```

```python
df_jif_q1.drop(['JCR Abbreviation','Publisher','Category','Edition','2023 JCI','% of OA Gold'], axis=1, inplace=True)
df_jif_q1.head()
```

```python
# Remover duplicatas na coluna 'Journal name'
df_jif_q1.drop_duplicates(subset=['Journal name'], inplace=True)

# Resetar o índice após remover os registros duplicados
df_jif_q1.reset_index(drop=True, inplace=True)

df_jif_q1.head()
```

```python
# Converter a coluna 'Total Citations' para numérico, removendo as vírgulas
df_jif_q1['Total Citations'] = df_jif_q1['Total Citations'].replace(',', '', regex=True)
df_jif_q1['Total Citations'] = pd.to_numeric(df_jif_q1['Total Citations'], errors='coerce')

# Converter a coluna '2023 JIF' para numérico, caso não esteja
df_jif_q1['2023 JIF'] = pd.to_numeric(df_jif_q1['2023 JIF'], errors='coerce')
```

```python
# Ordenar o DataFrame pelas colunas '2023 JIF' e 'Total Citations' em ordem decrescente
df_jif_q1.sort_values(by=['2023 JIF', 'Total Citations'], ascending=[False, False], inplace=True)

# Resetar o índice após a ordenação
df_jif_q1.reset_index(drop=True, inplace=True)

df_jif_q1.head()
```

```python
df_jif_q1.to_excel('../../data/processed/_6_journals-JIF-Q1.xlsx')
```
